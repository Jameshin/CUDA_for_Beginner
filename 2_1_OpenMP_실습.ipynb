```
code1 = """
// OpenMP program to print Hello World
// using C language

// OpenMP header
#include <omp.h>
#include <stdio.h>
#include <stdlib.h>

int main(int argc, char* argv[])
{
    #pragma omp parallel num_threads(4)
    {
        int my_rank = omp_get_thread_num();
        int thread_count = omp_get_num_threads();
        printf("Hello from thread %d of %d \\n", my_rank, thread_count);
    }
    return 0;
}
"""
```

```
code2 = '''
// OpenMP header
#include <stdio.h>
#include <stdlib.h>
#include <omp.h>
#include <iostream>

int main(int argc, char* argv[])
{
  int N = 1000;
  double* A = new double[N];
	double* B = new double[N];
  double* C_omp = new double[N];

  int thread_count = 4;

  #pragma omp parallel num_threads(thread_count)
  {
    int my_rank = omp_get_thread_num();
    int my_idx_start = my_rank * (N / thread_count);
    int my_idx_end = my_idx_start + (N / thread_count) - 1;

    for (int i = my_idx_start; i <= my_idx_end; i++)
    //#pragma omp parallel for
    //for (int i = 0; i < N; i++)
    {
      C_omp[i] = A[i] + B[i];
      //printf("Number of Threads: %d \\n", omp_get_num_threads());
    }

    printf("Thread %d computes from %d to %d\\n", my_rank, my_idx_start, my_idx_end);
  }

  return 0;
}
'''
```

```
code3 = '''
// OpenMP header
#include <omp.h>
#include <stdio.h>
#include <stdlib.h>
#include <omp.h>
#include <iostream>

int main(int argc, char* argv[])
{
  int N = 10000;
  double* A = new double[N];
	double* B = new double[N];
	double C = 0.0f;
  double C_omp = 0.0f;
  double temp;

  for (int i = 0; i < N; i++)
	{
		A[i] = 2.0;
		B[i] = 2.0;
	}

  //reduction(+:C_omp)   // private(temp)
  #pragma omp parallel for reduction(+:C_omp)
  for(int i=0; i<N; i++)
  {
    //temp = A[i] * B[i];
    //#pragma omp critical
    C_omp += A[i]*B[i];
  }

  //std::cout << "\\n Serial version : " << C << std::endl;
  std::cout << "\\n OpenMP for reduction 버전 : " << C_omp << std::endl;

  return 0;
}
'''
```
